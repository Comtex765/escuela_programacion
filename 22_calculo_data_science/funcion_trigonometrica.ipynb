```python
import matplotlib.pyplot as plt
import numpy as np
from math import pi
```

```python
def f(x):
  return np.sin(x)
```

```python
res = 100

x = np.linspace(-2*pi, 2*pi, num=res)

y = f(x)
```

```python
fig, ax = plt.subplots()


ax.plot(x,y)
ax.grid()
ax.axhline(y=0,color='r')
ax.axvline(x=0,color='r')
```
