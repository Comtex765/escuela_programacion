```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 2*x**8 - x**4 + 3*x**2 + 4
```

```python
res = 100

x = np.linspace(-2, 2, num=res)

y = f(x)
```

```python
fig, ax = plt.subplots()

plt.xlim(-1.5, 1.5)
plt.ylim(0, 100)
ax.plot(x,y)
ax.grid()
ax.axhline(y=0,color='r')
ax.axvline(x=0,color='r')
```
