```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(m, x, b):
    return m*x + b
```

```python
res = 100

m = 2

b = 1

x = np.linspace(-2, 2, num=res)

y = f(m,x,b)
```

```python
fig, ax = plt.subplots()


ax.plot(x,y)
ax.grid()
ax.axhline(y=0,color='r')
ax.axvline(x=0,color='r')
```
